## Trading View Report: 3 - Comparing Assets

Using normalized data, it becomes easier to compare assets.  Analysis is about understanding relationships, directions and intensity.
Because this is the case, we don't need to see the actual values to understand the behavior of the asset.  
A normalized representation can paint a similar picture that will tell us all we need to know.
This will become more useful if algorithms are applied to the data to make predictions about future events.

The same assets could be shown on one chart.  Assets from different markets can also be understood in a common context. 
Here two charts of different assets in the same market will be compared. 
The intent is to highlight that attributes can be evaluated by looking at assets moving in different directions.

In both cases the differences in close seems to have a similar range.
Big volume spikes seem to be more reflective of decreasing value than increase, but accompany more rapid changes than lower volume.
The downsloping case is more denslely distributed than the upslope.
It also seems to be the case that the difference between closes gets wider when the volume is higher.
The width is amplified in the upsloping example.
The normalization seems to enhance the perceived magnitude of big swings in volume, leaving the similar volumes to appear artificially low.

An important side note is that somewhere in the middle of the range the differnece will be 0.
This means that some of the lower values reflect a loss compared to previous close.
Bearing that in mind, the upsloping chart seems more volatile than the downsloping chart.

The observations presented here should not be considered to be definitive of trends.
Instead, they are presented as an entry point to further analysis, which may lead to an understanding of the relationship between price, volume and change from period to period.

Or, it may be the case that these attributes don't tell us what we want to know and can be replaced with other ones.
